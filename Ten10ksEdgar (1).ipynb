```python
import requests
from bs4 import BeautifulSoup
import pickle
import datetime
```

```python
# example search result page
r = requests.get('https://searchwww.sec.gov/EDGARFSClient/jsp/EDGAR_MainAccess.jsp?search_text=*&sort=Date&formType=Form10K&isAdv=true&stemming=true&numResults=100&fromDate=04/18/2019&toDate=04/18/2020&numResults=100')
```

```python
# print(soup.prettify())
```

```python
# results = set([])

# start with 04-17 and go backwards
end = datetime.date(2020, 4, 17)
start = end - datetime.timedelta(days=7)
```

```python
while(True):

    # search 1-week intervals
    fromDate = start.strftime('%m/%d/%Y')
    toDate = end.strftime('%m/%d/%Y')

    # EDGAR only returns max 8000 search results
    # TO DO: make 8100 a dynamic condition to save on repeat "last page searches"
    startDoc = 1
    while(startDoc < 8100):
        urlstring = "https://searchwww.sec.gov/EDGARFSClient/jsp/EDGAR_MainAccess.jsp?search_text=*&sort=Date&startDoc=" + str(startDoc) + "&numResults=100&isAdv=true&formType=Form10K&fromDate=" + fromDate + "&toDate=" + toDate + "&stemming=true"
        r = requests.get(urlstring)
        soup = BeautifulSoup(r.content, 'html.parser')
        tenks = soup.select('a[title = "Parent Filing"]')
        for tenk in tenks:
            link = str(tenk['href']).strip("javascript:opennewfiling('")
            results.add(link[:link.find(",")-1])
            # TO DO: extract filing date from results page
        startDoc += 100
        # print update to console
        if (len(results) % 50 <= 2): print(len(results))
        
    end = end - datetime.timedelta(days=7)
    start = start - datetime.timedelta(days=7)
    
    output = open('edgarlinks.pkl', 'wb')
    pickle.dump(results, output)
    output.close()
    
    print(len(results))
    print(end.strftime('%m/%d/%Y'))
    
    if end < datetime.date(2010, 1, 1): break
```
